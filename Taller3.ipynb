```python

## clase que contiene todo los metodos y atributos para la funcionalidad del algoritmo 

class Cinepython():
    
    
#definimos los datos a utilizar por eso creamos 3 diccionario 
#salas guardara los datos necesario de todas las salas para poder utilizar en el programa 

    salas = { 1: {'capacidad': 50, 'valor_hora':2500,'estado':'disponible'},
              2: {'capacidad': 40, 'valor_hora':2000, 'estado':'disponible'},
              3: {'capacidad': 60, 'valor_hora':3000,'estado':'disponible'},
              4: {'capacidad': 70, 'valor_hora':3500,'estado':'disponible'},
              5: {'capacidad': 80, 'valor_hora':4000,'estado':'disponible'},
              6: {'capacidad': 90, 'valor_hora':4500,'estado':'disponible'},
              7: {'capacidad': 100, 'valor_hora':5000,'estado':'disponible'}}
    
# menu guarda los datos de los menu que se ofrecen en Cinepython 
    menu = {1: {'item': 'hot dog', 'valor': 1500},  
            2: {'item': 'hamburguesa', 'valor': 2000},
            3: {'item': 'pizza', 'valor': 3000},
            4: {'item': 'combo: palomitas grande +gaseosa mediana ', 'valor': 5000},
            5: {'item': 'papas fritas', 'valor': 5000},
            6: {'item': 'pollo asado', 'valor': 10000}}
    
    
    menu_pedido={}
    

    
  #metodo Mostrar_sala nos permitira ver cuales son las salas que estan reservadas alquilada y disponibles 
    def Mostrar_salas(self, estado):
        
        llaves= self.salas.keys()
        x=""
        for item in llaves:
            
            if self.salas[item]["estado"]==estado:
                x= x+" \n Sala:"+str(item)+ "   Capacidad:" + str(self.salas[item]["capacidad"])+"  Costo Hora:"+str(self.salas[item]["valor_hora"])
        return x    
            
        
    # metodo Mostrar_menu que nos muestra la lista de comida que Cinepython tiene por ofrecer  
    def Mostrar_menu(self):
        llaves= self.menu.keys()
        x=""
        
        for item in llaves:
            
                x= x+" \n menu:"+str(item)+ "  :" + str(self.menu[item]["item"])+"  precio:"+str(self.menu[item]["valor"])
        return x 
        
        
   # metodo Alquilar_sala nos permitira alquilar cualquier sala que esta disponible y nos devovera el valor que cuesta esa sala segun las horas de alquilesr 
    def Alquilar_sala(self, num_sala , pelicula, nombre,hora,fecha_reserva):
        
        self.salas[num_sala]["usuario"]=nombre
        self.salas[num_sala]["serie_pelicula"]=pelicula
        self.salas[num_sala]["duracion_alquiler"]=hora
        self.salas[num_sala]["fecha_reserva"]=fecha_reserva
        self.salas[num_sala]["estado"]="alquilada"
     
        
    
    
    #metodo Reservar_sala nos permite reservar cualquier sala que esta disponible y nos debuelve el valor que cuetas la reserva de esa sala segun las horas rreservadas
    def Reservar_sala(self, num_sala, pelicula,nombre,hora,fecha_reserva):
        self.salas[num_sala]["usuario"]=nombre
        self.salas[num_sala]["serie_pelicula"]=pelicula
        self.salas[num_sala]["duracion_reserva"]=hora
        self.salas[num_sala]["fecha_reserva"]=fecha_reserva
        self.salas[num_sala]["estado"]="reservada"
        

        
    
    
    def Pedir_comida(self,cantidad=3, num_menu=1):
        
        costo_comida=0
        costo_comida=self.menu[num_menu]["valor"]*cantidad
        
        self.menu_pedido[num_menu]={}
        self.menu_pedido[num_menu]["item"]=self.menu[num_menu]["item"]
        self.menu_pedido[num_menu]["valor"]=self.menu[num_menu]["valor"]
        self.menu_pedido[num_menu]["cantidad_pedida"]=cantidad
        self.menu_pedido[num_menu]["precio_total"]=costo_comida
        
        
        return "Pedido Realizado Con Exito"
    
    # Metodo Factura que muestra y calcula los precios de los servicios 
    
    def Factura(self):
        llaves= self.salas.keys()
        llaves_menu=self.menu_pedido.keys()
        x=""
        costo_Total=0;
        
        
        for item in llaves:
            
            if self.salas[item]["estado"]=="reservada":
                costo_sala_r=self.salas[item]["duracion_reserva"]*self.salas[item]["valor_hora"]
               
                costo_Total=costo_Total+ costo_sala_r;
                x= x +" \n Sala:"+str(item)+ "   Capacidad:" + str(self.salas[item]["capacidad"])+"  Costo Hora:"+str(self.salas[item]["valor_hora"])+" Costo Total Sala: "+str(costo_sala_r)
            
            elif self.salas[item]["estado"]=="alquilada":
                costo_sala_a=self.salas[item]["duracion_alquiler"]*self.salas[item]["valor_hora"]*self.salas[item]["capacidad"]
                x= x +"  Sala:"+str(item)+ "   Capacidad:" + str(self.salas[item]["capacidad"])+"  Costo Hora:"+str(self.salas[item]["valor_hora"])+" Costo Total Sala: "+str(costo_sala_a)
                costo_Total=costo_Total+ costo_sala_a;
                
                
        for item in llaves_menu:
            x= x +" \n Menu:"+self.menu_pedido[item]["item"]+" Precio Unidad:"+str(self.menu_pedido[item]["valor"])+" Cantidad Pedida:"+str(self.menu_pedido[item]["cantidad_pedida"])+" Pago:"+str(self.menu_pedido[item]["precio_total"])
            costo_Total=costo_Total + self.menu_pedido[item]["precio_total"]
            
                    
                    
        return x +""" \n Costo total neto $"""+str(costo_Total)    
        
objeto = Cinepython()
```
